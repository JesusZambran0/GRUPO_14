Notebook base para inspeccionar datasets públicos de YouTube colocados en `data/raw/`.

```python
from pathlib import Path
import pandas as pd
import numpy as np

RAW = Path("../data/raw")
files = sorted(RAW.glob("*.csv"))
files
```

```python
frames = []
for f in files:
    df = pd.read_csv(f, low_memory=False)
    df["source_file"] = f.name
    frames.append(df)
raw = pd.concat(frames, ignore_index=True, sort=False) if frames else pd.DataFrame()
raw.shape
```
